```python

x=np.linspace(0,6,201)###we consider this interval with 201 points to make our drawing nicer
y=np.linspace(0,4,201)###we do same here
t_plate=lambda x,y: 100*np.exp(-0.4*((x-1)**2+0.7*(y-3)**2))+80*np.exp(-0.2*(2*(x-5)**2+1.5*(y-1)**2))###used the lambda function given in the question 
X,Y=np.meshgrid(x,y)###formed ordered pair with this 
F=t_plate(X,Y)##stored the ordered pair values in the capital F
plt.figure(figsize=(10,8))
plt.contourf(X,Y,F)## used the python's inbuilt function to make a contour plate 
```

```python
###bi###
def cool(plate,kernall):
    m,n=plate.shape
    result=np.zeros_like(plate)
    for x in range(1,m-1):
        for y in range(1,n-1):
            result[x,y]=(kernal*plate[x-1:x+2,y-1:y+2]).sum()
    return result        
```

```python
kernal=np.ones((3,3))*(1/8)
kernal

#creating a 3X3 kernel
```

```python
pixelaverage  =  cool(F,kernal)
plt.contourf(X,Y,pixelaverage)
```

```python
#bii(outer product of a level 2 pascals triangle)
a=np.array([1,2,1])#coefficients of level two pascals triangle
b=np.outer(a,a)#we find the outer product by using numpy's inbuilt function which is the outer to provide us with the matrix multiplication
b
#note, the outer product just multiplies the entries in the first arrray by each entry in the next array, accordingly.
```

```python
###biii###
a = cool(F,a)
plt.contourf(X,Y,a)
```
