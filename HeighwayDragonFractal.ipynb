```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(z):
    return ((1.+1j)*z)/2.
def f2(z):
    return 1.-((1.-1j)*z)/2.
```

```python
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)
z = np.sin(xx**4 + yy**1) / (xx**2 + yy**5)
plt.figure(figsize=(10,8)) 
h = plt.contourf(x, y, z)
plt.axis('scaled')
plt.show()
```

```python
def dragon(n):
    s=[0,1]
    zvals=np.array(s)
    for i in range(n):
        zvals=(np.hstack((f1(zvals),f2(zvals[-1::-1]))))
    R_zvals=zvals.real 
    I_zvals=zvals.imag
    plt.plot(R_zvals,I_zvals,linewidth=3)
    
    # g=plt.gca().spines
    # gtop=g['top'].set_color('none')
    # gbottom=g['bottom'].set_position(('data',0))
    # gright=g['right'].set_color('none')
    # gleft=g['left'].set_position(('data',0))
```

```python

plt.figure(figsize=(25,10))
plt.subplot(2,5,1)
dragon(0)
plt.title('S0',color='m')
plt.subplot(2,5,2)
dragon(1)
plt.title('S1',color='m')
plt.subplot(2,5,3)
dragon(2)
plt.title('S2',color='m')
plt.subplot(2,5,4)
dragon(3)
plt.title('S3',color='m')
plt.subplot(2,5,5)
dragon(4)
plt.title('S4',color='m')
plt.subplot(2,5,6)
dragon(5)
plt.title('S5',color='m')
plt.subplot(2,5,7)
dragon(6)
plt.title('S6',color='m')
plt.subplot(2,5,8)
dragon(7)
plt.title('S7',color='m')
plt.subplot(2,5,9)
dragon(8)
plt.title('S8',color='m')
```

```python
plt.figure(figsize=(20,20))
dragon(12)
plt.title('Random',color='m')
plt.show()
```

```python
def f1b(z):
    return ((1.+1j)*z)/2.
def f2b(z):
    return 1.-((1.+1j)*z)/2.
```

```python
def twindrag(n):
    s=[0,1, 1-1j]
    zvals=np.array(s)
    for i in range(n):
        zvals=(np.hstack((f1b(zvals),f2b(zvals[-1::-1]))))
    R_zvals=zvals.real 
    I_zvals=zvals.imag
    plt.plot(R_zvals,I_zvals,linewidth=3)
```

```python
plt.figure(figsize=(25,10))
plt.subplot(2,5,1)
twindrag(0)
plt.title('S0',color='m')
plt.subplot(2,5,2)
twindrag(1)
```

```python
plt.figure(figsize=(20,20))
twindrag(12)
plt.title('Random',color='m')
plt.show()
```

```python
for i in range(2):
    plt.figure(figsize=(10,10))
    twindrag(12*i)
    plt.title('Random',color='m')
    plt.show()
```

```python
from math import sqrt
```

```python
def t1(z):
    return (1./2. - (1j/(2.*sqrt(3.))))*z
def t2(z):
    return ((1j/sqrt(3.)) *z) + (1./2. -1j/(2.*sqrt(3.)))
def t3(z):
    return (1./2. - (1j/(2.*sqrt(3.))))*z + (1./2. +1j/(2.*sqrt(3.)))
```

```python
def tdragon(n):
    s=[0.,1]
    zvals=np.array(s)
    # zvals2=np.array(s)
    # zvals3=np.array(s)
    for i in range(n):
        zvals=(np.hstack((t1(zvals),t2(zvals), t3(zvals))))
        # zvals2=(np.hstack((t2(zvals),t3(zvals[-1::-1]))))
        # zvals3=(np.hstack((t1(zvals),t2(zvals[-1::-1]))))
    R_zvals=zvals.real 
    I_zvals=zvals.imag
    # R_zvals2=zvals2.real 
    # I_zvals2=zvals2.imag 
    # R_zvals3=zvals3.real 
    # I_zvals3=zvals3.imag
    plt.plot(R_zvals,I_zvals,color='b',linewidth=3)
    # plt.plot(R_zvals2,I_zvals2,linewidth=3)
    # plt.plot(R_zvals3,I_zvals3,linewidth=3)
```

```python
for i in range(4):
    plt.figure(figsize=(15,15))
    tdragon(4*i)
    plt.title('Random',color='m')
    plt.show()
```
